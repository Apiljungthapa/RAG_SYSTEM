# 4. Loading the model & defining prompt for summarizing texts

```python

load_dotenv()

gemini_api_key = os.getenv("GEMINI_API_KEY")
pinecone_api = os.getenv("PINECONE_API")

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key=gemini_api_key,
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    
)
```

```python
prompt_text = """
You are a helpful assistant that summarizes content extracted from business documents, including both text and tables.

Your goal is to produce clear, concise, and informative summaries by:
- Extracting the most important insights from the input (whether textual, tabular, or a mix of both).
- Combining numerical and narrative data when needed to give a complete picture.
- Avoiding any introduction, explanation, or phrases like "Here is the summary" or "Summary:".
- Writing in a neutral, professional tone, suitable for business reporting.

Respond with only the summary.

Input: {element}
"""

prompt = ChatPromptTemplate.from_template(prompt_text)

summarize_chain = {"element": lambda x: x} | prompt | llm | StrOutputParser()

summary_dict = {}

def summarize_texts(texts):
    for idx, text in enumerate(texts):
        print(f"\nText {idx}:")
        print(text.text)
        
        summary = summarize_chain.invoke({"element": text.text})
        key = f"summary_{idx}"
        
        summary_dict[key] = summary

    return summary_dict

```

```python
summarize_texts(texts)
```

# 6. Querying Pinecone DB

```python
query = "What were Apple Inc.'s net sales figures from product and service for iPhone, Mac, iPad, Wearables, Home and Accessories, and Services for the three months ended June 25, 2022?"

# Get dense vector embedding
query_embedding = model.encode([query])[0] 
```

# 7. Defining prompt for answer summarization into format

```python
prompt_text = """
You are a highly accurate and concise AI assistant.

You are provided with:
1. A user’s factual question.
2. The full original retrieved text relevant to the question.
3. A concise summary of that text.

Your tasks:
- Understand the user’s question carefully.
- Use both the original text and its summary to extract exact factual information (e.g., dates, figures, names).
- Resolve ambiguity if present using reasoning and cross-referencing.
- Respond with a complete, precise sentence that directly answers the question.
- Your answer must be naturally phrased — as a full, clear, and informative sentence — not just a fragment, number, or date.

Output format:
- A single complete and accurate sentence answering the question directly.
- No preamble, explanation, repetition, or extra commentary.

---

**Question:** {query}

**Original Retrieved Text:** {text}

**Summary of the Text:** {summary}

---

Respond with one complete, factual sentence only.
"""


prompt = PromptTemplate(
    input_variables=["query", "text", "summary"],
    template=prompt_text.strip()
)

output_parser = StrOutputParser()

summarize_chain = (
    RunnableMap({
        "query": lambda x: x["query"],
        "text": lambda x: x["text"],
        "summary": lambda x: x["summary"]
    }) 
    | prompt 
    | llm 
    | output_parser
)

```

```python
response = summarize_chain.invoke({
    "query": "What were Apple Inc.'s net sales figures from product and service for iPhone, Mac, iPad, Wearables, Home and Accessories, and Services for the three months ended June 25, 2022?",
    "text": results["matches"][0]["metadata"]["text"],
    "summary": results["matches"][0]["metadata"]["summary"]
})

print(response)
```
